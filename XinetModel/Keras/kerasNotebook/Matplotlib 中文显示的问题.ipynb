# 第一种方法

```python
from pylab import mpl
import numpy as np

mpl.rcParams['font.sans-serif'] = ['SimHei'] # 指定默认字体  
  
mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号 '-' 显示为方块的问题  
%pylab inline
```

```python
t = np.arange(-5 * np.pi, 5 * np.pi, 0.01)  
y = np.sin(t)/t  
plt.plot(t, y)  
plt.title(u'这里写的是中文')  
plt.xlabel(u'X坐标')  
plt.ylabel(u'Y坐标')  
plt.show()  
```
